## Scatterplot

For creating scatterplots, qplot() is straightforward to use. Here, we specify:
- data frame containing the source data
- independent and dependent variables
- color the plot by a categorical (factor) variable, or by a numeric variable
- labels for the axes

```r
plotObj <- qplot(data=diamonds, 
                 carat, price, 
                 color=clarity, 
                 xlab="Carats", 
                 ylab="Price in Dollars", 
                 main="Diamond Price with Size")
```

This code allows scaling of the plot size. Call it with a width and height arguments specified in the cell where you print the plot.

```r
fig <- function(width, height){
     options(repr.plot.width = width, repr.plot.height = height)
}
# per this reference: https://www.kaggle.com/getting-started/105201
```

Specify the plot name to print it to the screen:

```r
fig(10,6)
plotObj
```

## Faceted barplot

```r
cutObj <- ggplot(diamonds, aes(color)) +
             geom_bar(stat="count", aes(fill = factor(cut))) + 
             facet_grid(~cut) 
```

```r
# Make this faceted ggplot more readable with a call to the fig function defined above:
fig(10,4)
cutObj
```
